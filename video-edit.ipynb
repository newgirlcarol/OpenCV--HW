```python
# Creat a video with different funtion
import cv2
import numpy as np
import dlib
# load video
fourcc = cv2.VideoWriter_fourcc(*'MGPG')
cap = cv2.VideoCapture('./video/HP1.mp4')
# print(f'frame_w : {cap.get(cv2.CAP_PROP_FRAME_WIDTH)}\t frame_h :{cap.get(cv2.CAP_PROP_FRAME_HEIGHT)}')
# print(f'frame_fps : {cap.get(cv2.CAP_PROP_FPS)}')
# print(f'frame_count : {cap.get(cv2.CAP_PROP_FRAME_COUNT)}')
fgbg = cv2.bgsegm.createBackgroundSubtractorMOG(history=200, nmixtures=5, backgroundRatio=0.7)
mog = cv2.cv2.createBackgroundSubtractorMOG2(history=200, detectShadows = False)
bs = cv2.createBackgroundSubtractorKNN(detectShadows=True)
tracker = cv2.TrackerMIL_create()

detector = dlib.get_frontal_face_detector()
out = cv2.VideoWriter('./video/HP_out1.mp4', fourcc, 33.0, (800, 450))
#--------------------------------------------------------------
# Sobel
while (1):
    ret, frame = cap.read()
    if not ret: break
    frame = cv2.resize(frame, (800, 450), interpolation = cv2.INTER_NEAREST)
    
    sobelx = cv2.Sobel(frame, cv2.CV_64F, 1, 0, ksize=3)
    sobely = cv2.Sobel(frame, cv2.CV_64F, 0, 1, ksize=3)
    sobelx = cv2.convertScaleAbs(sobelx)   # 轉回 Uint8 
    sobely = cv2.convertScaleAbs(sobely)  
    frame =  cv2.addWeighted(sobelx, 0.5, sobely, 0.5, 0)
    cv2.putText(frame, '1.Sobel', (550, 420), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (11,255,255), 2)
    cv2.imshow('frame',frame)
    out.write(frame)

    if 8100 <= int(cap.get(cv2.CAP_PROP_POS_MSEC)) <= 8300:
        break
    k = cv2.waitKey(1) & 0xff
    if k == 27:
        break
#--------------------------------------------------------------
while (1):
    ret, frame = cap.read()
    if not ret: break
    frame = cv2.resize(frame, (800, 450), interpolation = cv2.INTER_NEAREST)
    cv2.imshow('frame',frame)
    out.write(frame)
#     print(int(cap.get(cv2.CAP_PROP_POS_MSEC)))
    if 10400 <= int(cap.get(cv2.CAP_PROP_POS_MSEC)) <= 10500:
        break
    k = cv2.waitKey(1) & 0xff
    if k == 27:
        break
#--------------------------------------------------------------
# MOG 背景提取
#Canny 特效:apertureSize:3 >>線太少 >> 改為 5|7, threshold2調到450
while (1):
    ret, frame = cap.read()
    if not ret: break
    frame = cv2.resize(frame, (800, 450), interpolation = cv2.INTER_NEAREST)
    frame = mog.apply(frame)
    cv2.putText(frame, '2.MOG', (550, 420), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (255,255,255), 2)
    cv2.imshow('frame',frame)
    out.write(frame)
    if 21600 <= int(cap.get(cv2.CAP_PROP_POS_MSEC)) <= 21700:
        break      
    k = cv2.waitKey(1) & 0xff
    if k == 27:
        break
#--------------------------------------------------------------
while (1):
    ret, frame = cap.read()
    if not ret: break
    frame = cv2.resize(frame, (800, 450), interpolation = cv2.INTER_NEAREST)
    cv2.imshow('frame',frame)
    out.write(frame)
#     print(int(cap.get(cv2.CAP_PROP_POS_MSEC)))
    if 24000 <= int(cap.get(cv2.CAP_PROP_POS_MSEC)) <= 24100:
        break
    k = cv2.waitKey(1) & 0xff
    if k == 27:
        break
#--------------------------------------------------------------
while (1):
    ret, frame = cap.read()
    if not ret: break
    frame = cv2.resize(frame, (800, 450), interpolation = cv2.INTER_NEAREST)
    frame = cv2.Canny(frame, 30, 150, apertureSize=3, L2gradient=0)
    cv2.putText(frame, '3.Canny', (550, 420), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (255,255,255), 2)
    cv2.imshow('frame',frame)
    out.write(frame)
#     print(int(cap.get(cv2.CAP_PROP_POS_MSEC)))
    if 45000 <= int(cap.get(cv2.CAP_PROP_POS_MSEC)) <= 46000:
        break      
    k = cv2.waitKey(1) & 0xff
    if k == 27:
        break
#--------------------------------------------------------------
while (1):
    ret, frame = cap.read()
    if not ret: break
    frame = cv2.resize(frame, (800, 450), interpolation = cv2.INTER_NEAREST)
    cv2.imshow('frame',frame)
    out.write(frame)
#     print(int(cap.get(cv2.CAP_PROP_POS_MSEC)))
    if 50000 <= int(cap.get(cv2.CAP_PROP_POS_MSEC)) <= 51500:
        break
    k = cv2.waitKey(1) & 0xff
    if k == 27:
        break
#--------------------------------------------------------------
while (cap.isOpened()):
    ret, frame = cap.read()
    frame = cv2.resize(frame, (800, 450), interpolation = cv2.INTER_NEAREST)
    if ret :
        face_rects, scores, idx = detector.run(frame, 0, -.5) 
        for i, d in enumerate(face_rects):               # 取出所有偵測的結果
            x1 = d.left(); y1 = d.top(); x2 = d.right(); y2 = d.bottom()
            text = f'{scores[i]:.2f}, ({idx[i]:0.0f})'

            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 4, cv2.LINE_AA) # 以方框標示偵測的人臉
            cv2.putText(frame, text, (x1, y1), cv2.FONT_HERSHEY_DUPLEX,          # 標示分數
                        0.7, (255, 255, 255), 1, cv2.LINE_AA)
    cv2.putText(frame, '4.Face Detection', (550, 420), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (11,255,255), 2)
    out.write(frame)
    cv2.imshow('frame',frame)
#     print(int(cap.get(cv2.CAP_PROP_POS_MSEC)))
    if 57300 <= int(cap.get(cv2.CAP_PROP_POS_MSEC)) <= 57800:
        break       
    k = cv2.waitKey(1) & 0xff
    if k == 27:
        break
#--------------------------------------------------------------
while (1):
    ret, frame = cap.read()
    if not ret: break
    frame = cv2.resize(frame, (800, 450), interpolation = cv2.INTER_NEAREST)
    cv2.imshow('frame',frame)
    out.write(frame)
#     print(int(cap.get(cv2.CAP_PROP_POS_MSEC)))
    if 58000 <= int(cap.get(cv2.CAP_PROP_POS_MSEC)) <= 58500:
        break
    k = cv2.waitKey(1) & 0xff
    if k == 27:
        break
# while (1):
#     ret, frame = cap.read()
#     if not ret: break
#     frame = cv2.resize(frame, (800, 450), interpolation = cv2.INTER_NEAREST)
#     blurred = cv2.medianBlur(frame, 7)
#     flipped = cv2.flip(blurred, 0)
#     cv2.putText(flipped, '5.Blur+Flip', (550, 420), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (255,255,255), 2)
#     cv2.imshow('frame',flipped)
#     if 60000 < (cap.get(cv2.CAP_PROP_FPS)) < 60100:
#         break
        
#     k = cv2.waitKey(1) & 0xff
#     if k == 27:
#         break
#--------------------------------------------------------------
while (1):
    ret, frame = cap.read()
    if not ret: break
    frame = cv2.resize(frame, (800, 450), interpolation = cv2.INTER_NEAREST)  
    frame = bs.apply(frame)
    cv2.putText(frame, '5.KNN', (550, 420), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (255,255,255), 2)
    out.write(frame)
    cv2.imshow('frame',frame)        
    k = cv2.waitKey(1) & 0xff
    if k == 27:
        break

cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
